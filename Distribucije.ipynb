# 1-D

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
N, M = 100000, 1000
x = np.random.binomial(1, 0.5, size=(N,M))
x[x==0] = -1

traces = np.cumsum(x, axis=0)/1000
r = traces[-1,:]
loc, scale = norm.fit(r)
print(f'Normalna distribucija sa eta={loc:.2f} i sigma={scale:.2f}')

x = np.linspace(-5, 5, 100)
plt.figure(figsize=[15, 7])
plt.plot(x, norm.pdf(x, loc=loc, scale=scale), 'r-', lw=2, alpha=0.6, label='Normalna distribucija')
plt.hist(r, density=True, alpha = 0.6, label='Histogram 1D slučajne šetnje')
plt.xlabel('$S_n$'); plt.ylabel('$p(S_n)$')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend();
```

# 2-D

```python
from scipy.stats import norm, rayleigh, chi
import numpy as np
import matplotlib.pyplot as plt

N, M = 100000, 1000

x = np.random.binomial(1, 0.5, size=(N, M))
x[x==0] = -1
y = np.random.binomial(1, 0.5, size=(N, M))
y[y==0] = -1
r = ((np.sum(x, axis=0)/1000)**2 + (np.sum(y, axis=0)/1000)**2)**0.5

loc, scale = rayleigh.fit(r)
print(f'Rayleigh distribucija sa eta={loc:.2f} i sigma={scale:.2f}')
dof, loc, scale = chi.fit(r)
print(f'Chi distribucija sa dof={dof:.2f}, eta={loc:.2f} i sigma={scale:.2f}')
```

```python
x = np.linspace(0,5, 100)
plt.figure(figsize=[15, 7])
plt.plot(x, rayleigh.pdf(x, loc=loc, scale=scale), 'r-', lw=2, alpha=0.6, label='Rayleigh distribucija')
plt.scatter(x, chi.pdf(x, 2, loc=loc, scale=scale), alpha=0.6, c='k', label ='Chi distribucija (k=2)')
plt.hist(r,density=True, alpha=0.5, label='Histogram 2D slučajne šetnje')
plt.xlabel('$S_n$'); plt.ylabel('$p(S_n)$')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend();
```

# 3-D

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

N, M = 100000, 1000

x = np.random.binomial(1, 0.5, size=(N, M))
x[x==0] = -1
y = np.random.binomial(1, 0.5, size=(N, M))
y[y==0] = -1
z = np.random.binomial(1, 0.5, size=(N, M))
z[z==0] = -1
r = ((np.sum(x, axis=0)/1000)**2 + (np.sum(y, axis=0)/1000)**2 + (np.sum(z, axis=0)/1000)**2)**0.5

from scipy.stats import maxwell, chi
loc, scale = maxwell.fit(r)
print(f'Maxwell distribucija sa eta={loc:.2f} i sigma={scale:.2f}')
dof, loc, scale = chi.fit(r)
print(f'Chi distribucija sa k={dof:.2f}, eta={loc:.2f} i sigma={scale:.2f}')
```

```python
x = np.linspace(0,5, 100)
plt.figure(figsize=[15, 7])
plt.plot(x, maxwell.pdf(x, loc=loc, scale=scale), 'r-', lw=2, alpha=0.6, label='Maxwell distribucija')
plt.scatter(x, chi.pdf(x, 3, loc=loc, scale=scale), alpha=0.6, c='k', label ='Chi distribucija (k=3)')
plt.hist(r, bins=20, density=True, alpha=0.5, label='Histogram 3D slučajne šetnje')
plt.xlabel('$S_n$'); plt.ylabel('$p(S_n)$')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend();
```
