```python
import numpy as np
import matplotlib.pyplot as plt

def RandomWalk1D(N, p):
    y = np.empty(N+1, dtype=int)
    y[0] = 0
    x = np.random.choice([-1, 1], size=N, p=p)

    for i in range(N):
        y[i+1] = y[i] + x[i]

    return y, x

N = 20
p = [0.5, 0.5]
y, x = RandomWalk1D(N, p)

plt.figure(figsize=(8, 5))
plt.stem(range(1, N+1), x, linefmt='k--', markerfmt='ko', basefmt='k')
plt.xlabel(r'$n$')
plt.ylabel(r'$X_n$')
plt.xticks(range(1, N+1))
plt.yticks([-1, 0, 1])
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(8, 5))
plt.stem(range(N+1), y, linefmt='k--', markerfmt='ko', basefmt='k')
plt.xlabel(r'$n$')
plt.ylabel(r'$S_n$')
plt.xticks(range(0, N+1, 2))
plt.yticks(np.arange(min(y)-1, max(y)+1))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def RandomWalk1D(N, p):
  y = np.empty(N+1);
  y[0];
  for i in range(N):
    y[i+1] = y[i] + np.random.choice([-1, 1], p = p);

  return y

N = 500;
p = [0.5, 0.5];
plt.figure(figsize=[12,8])
for i in range(10):
  plt.plot(range(N+1), RandomWalk1D(N, p))

plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel(r'$n$')
plt.ylabel(r'$S_n$')
plt.show()

N = 500;
p = [0.6, 0.4];

plt.figure(figsize=[12,8])
for i in range(10):
  plt.plot(range(N+1), RandomWalk1D(N, p))

plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel(r'$n$')
plt.ylabel(r'$S_n$')
plt.show()
```

```python
import numpy as np
import random

n = 1000
x = np.zeros(n)
y = np.zeros(n)
direction=["NORTH","SOUTH","EAST","WEST"]
for i in range(1, n):
    step = random.choice(direction)
    if step == "EAST":
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
    elif step == "WEST":
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
    elif step == "NORTH":
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1
    else:
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1
plt.figure(figsize=[10,10])
plt.plot(x, y)
plt.plot(x[n-1], y[n-1], 'o', color='red')
plt.plot(x[0], y[0], 'o', color='green')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel('X'); plt.ylabel('Y')
plt.show()
```
