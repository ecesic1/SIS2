```python
import numpy as np
import matplotlib.pyplot as plt

n_steps = 10000
n_sim = 3

cases = {
    "E[X1] > 0": np.random.choice([1, 2], size=(n_sim, n_steps)),
    "E[X1] < 0": np.random.choice([-2, -1], size=(n_sim, n_steps)),
    "E[X1] = 0": np.random.choice([-1, 1], size=(n_sim, n_steps))
}

plt.figure(figsize=(12, 6))

for label, walks in cases.items():
    mean_val = np.mean(walks)
    label2 = f"{label} (E[X1] ≈ {mean_val:.2f})"
    first = True

    for walk in walks:
        S_n = np.cumsum(walk)
        brzina = S_n / np.arange(1, n_steps + 1)
        plt.plot(brzina, label=label2 if first else "")
        first = False

plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0, color='black', linestyle='--', linewidth=1)
plt.xlabel("Koraci n")
plt.ylabel("Asimptotska brzina šetnje")
plt.legend()
plt.show()
```
